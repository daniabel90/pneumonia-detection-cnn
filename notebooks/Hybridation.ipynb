```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2, EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Input, concatenate, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import os

# 1. Paramètres
img_size = (224, 224)
batch_size = 32
num_classes = 2  # ⚠️ Modifie ici si tu as 3 classes

# 2. Dossiers du dataset
train_dir = 'D:/Data-Chest-ray/train'
val_dir = 'D:/Data-Chest-ray/val'
test_dir = 'D:/Data-Chest-ray/test'

# 3. Chargement des données robustes (et labels one-hot)
train_data = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'  # one-hot encoding
)

val_data = tf.keras.utils.image_dataset_from_directory(
    val_dir,
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)

# 4. Prétraitement et performances
AUTOTUNE = tf.data.AUTOTUNE
train_data = train_data.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_data = val_data.cache().prefetch(buffer_size=AUTOTUNE)

# 5. Modèles de base sans leur top
input_tensor = Input(shape=(224, 224, 3))

base_model1 = MobileNetV2(weights='imagenet', include_top=False, input_tensor=input_tensor)
base_model2 = EfficientNetB0(weights='imagenet', include_top=False, input_tensor=input_tensor)

# 6. Geler les poids
base_model1.trainable = False
base_model2.trainable = False

# 7. Extraction des features
x1 = base_model1.output
x1 = GlobalAveragePooling2D()(x1)

x2 = base_model2.output
x2 = GlobalAveragePooling2D()(x2)

# 8. Fusion + couches personnalisées
x = concatenate([x1, x2])
x = Dense(256, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

# 9. Modèle final
hybrid_model = Model(inputs=input_tensor, outputs=output)
hybrid_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
hybrid_model.summary()

# 10. Callback EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# 11. Entraînement
history = hybrid_model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    callbacks=[early_stop]
)

# 12. Courbes d’apprentissage
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.title("Courbes d'entraînement")
plt.xlabel('Épochs')
plt.ylabel('Valeur')
plt.grid()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import seaborn as sns

```

```python
# 13. Prédictions sur validation
y_pred_probs = hybrid_model.predict(val_data)
y_pred = np.argmax(y_pred_probs, axis=1)

# 14. Vrais labels
y_true = np.concatenate([np.argmax(y, axis=1) for _, y in val_data], axis=0)

# 15. Matrice de confusion
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Matrice de confusion")
plt.xlabel("Prédiction")
plt.ylabel("Réel")
plt.show()

# 16. Rapport de classification (avec F1-score, etc.)
print("Classification Report:\n")
print(classification_report(y_true, y_pred))
```

```python
# 17. Courbes ROC
fpr = {}
tpr = {}
roc_auc = {}
num_classes = y_pred_probs.shape[1]

y_true_bin = tf.keras.utils.to_categorical(y_true, num_classes=num_classes)

plt.figure(figsize=(8, 6))
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'Classe {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbes ROC par classe')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50, MobileNetV2
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import os

# 1. Paramètres
img_size = (224, 224)
batch_size = 32
num_classes = 2  # ajustez selon vos classes

# 2. Prétraitement des données
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True
)
val_test_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory("D:/Data-Chest-ray/train", target_size=img_size, batch_size=batch_size, class_mode="categorical")
val_data = val_test_datagen.flow_from_directory("D:/Data-Chest-ray/val", target_size=img_size, batch_size=batch_size, class_mode="categorical")
test_data = val_test_datagen.flow_from_directory("D:/Data-Chest-ray/test", target_size=img_size, batch_size=batch_size, class_mode="categorical", shuffle=False)

# 3. Entrée commune
input_tensor = Input(shape=(224, 224, 3))

# 4. Modèle ResNet50
resnet = ResNet50(include_top=False, weights='imagenet', input_tensor=input_tensor)
for layer in resnet.layers:
    layer._name = "resnet_" + layer.name
resnet.trainable = False
resnet_output = GlobalAveragePooling2D()(resnet.output)

# 5. Modèle MobileNetV2
mobilenet = MobileNetV2(include_top=False, weights='imagenet', input_tensor=input_tensor)
for layer in mobilenet.layers:
    layer._name = "mobilenet_" + layer.name
mobilenet.trainable = False
mobilenet_output = GlobalAveragePooling2D()(mobilenet.output)

# 6. Fusion + Classifieur personnalisé
merged = Concatenate()([resnet_output, mobilenet_output])
x = Dense(256, activation='relu')(merged)
x = Dropout(0.2)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

# 7. Modèle final
hybrid_model = Model(inputs=input_tensor, outputs=output)
hybrid_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
hybrid_model.summary()

# 8. Callback
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# 9. Entraînement
history = hybrid_model.fit(train_data, validation_data=val_data, epochs=20, callbacks=[early_stop])

# 10. Courbes d’apprentissage
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Courbe d'apprentissage")
plt.show()

# 11. Évaluation
test_data.reset()
predictions = hybrid_model.predict(test_data)
y_pred = np.argmax(predictions, axis=1)
y_true = test_data.classes

# 12. Rapport de classification
print("Classification Report:\n", classification_report(y_true, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

# 13. Courbes ROC
fpr = dict()
tpr = dict()
roc_auc = dict()
y_test = test_data.classes
y_one_hot = tf.keras.utils.to_categorical(y_test, num_classes)

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_one_hot[:, i], predictions[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label=f'Classe {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.title('Courbes ROC')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
