```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Chemin racine du dataset
base_dir = 'C:/Users/Bsi/OneDrive/Bureau/Dania/Data-Chest-ray/Data-Chest-ray'  # ⚠️ Remplace par le chemin réel

# Sous-dossiers
splits = ['train', 'val', 'test']
classes = ['Normal', 'Pneumonia']

# Initialisation du compteur
counts = {split: {cls: 0 for cls in classes} for split in splits}

# Parcours de chaque split et classe
for split in splits:
    for cls in classes:
        folder = os.path.join(base_dir, split, cls)
        if os.path.exists(folder):
            counts[split][cls] = len([
                f for f in os.listdir(folder)
                if f.lower().endswith(('.png', '.jpg', '.jpeg'))
            ])
        else:
            print(f"⚠️ Dossier introuvable : {folder}")

# Affichage de l'histogramme
x = np.arange(len(classes))  # Positions des classes
width = 0.25  # Largeur des barres

fig, ax = plt.subplots(figsize=(10, 6))
for i, split in enumerate(splits):
    values = [counts[split][cls] for cls in classes]
    ax.bar(x + i * width, values, width=width, label=split)

# Configuration du graphique
ax.set_xlabel('Classe')
ax.set_ylabel('Nombre d\'images')
ax.set_title('Répartition des images par classe et par ensemble')
ax.set_xticks(x + width)
ax.set_xticklabels(classes)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Somme globale des images par classe
total_counts = {cls: sum(counts[split][cls] for split in splits) for cls in classes}

# Données pour le camembert
labels = list(total_counts.keys())
sizes = list(total_counts.values())
colors = ['#66b3ff', '#ff9999']  # Couleurs pour Normal et Pneumonia
explode = (0.05, 0.05)  # Décalage pour mettre en valeur les parts

# Affichage du camembert
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%',
        startangle=140, shadow=True)
plt.title('Répartition globale des images (Normal vs Pneumonia)')
plt.axis('equal')  # Pour un cercle parfait
plt.show()
```
